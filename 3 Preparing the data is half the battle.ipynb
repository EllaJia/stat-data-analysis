```python
# Cleaning Data

# Calculating and Removing Outliers
'''
    How to detect outliers:
        - Standard Deviations: If the data is normally distributed, then 95 percent of the data is within 1.96 standard deviations of the mean.
        - Interquartile Range (IQR): The IQR is the difference between the 25 percent quantile and the 75 percent quantile.
'''
```

```python
# Standard Deviations
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
meangrade = df['grade'].mean()
stdgrade = df['grade'].std()
toprange = meangrade + stdgrade * 1.96
botrange = meangrade - stdgrade * 1.96
copydf = df
copydf = copydf.drop(copydf[copydf['grade'] > toprange].index)
copydf = copydf.drop(copydf[copydf['grade'] < botrange].index)
copydf
```

```python
# Interquartile Range
import pandas as pd
Location = 'datasets/gradedata.csv'
df = pd.read_csv(Location)
q1 = df['grade'].quantile(.25)
q3 = df['grade'].quantile(.75)
iqr = q3 - q1
toprange = q3 + iqr * 1.5
botrange = q1 - iqr * 1.5
copydf = df
copydf = copydf.drop(copydf[copydf['grade'] > toprange].index)
copydf = copydf.drop(copydf[copydf['grade'] < botrange].index)
copydf
```

```python
# Removing Punctuation from Column Contents
import pandas as pd
Location = 'datasets/gradedata.csv'
df = pd.read_csv(Location)
df.head()
```

```python
import string
exclude = set(string.punctuation)
def remove_punctuation(x):
    try:
        x = ''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x
df.address = df.address.apply(remove_punctuation) # you can use df.address to call column "address" in df
df
```

```python
# Removing Whitespace from Column Contents
import pandas as pd
Location = 'datasets/gradedata.csv'
df = pd.read_csv(Location)
df.head()
```

```python
# standardizing text like SSNs, phone numbers, and zip codes
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]
ssns = ['867-53-0909','333-22-4444','123-12-1234','777-93-9311','123-12-1423']
GradeList = zip(names,grades,bsdegrees,msdegrees,phddegrees,ssns)
columns=['Names','Grades','BS','MS','PhD',"ssn"]
df = pd.DataFrame(data = GradeList, columns=columns)
df
```

```python
# Remove hyphens from SSNs and add leading zeros if necessary
def right(s, amount):
    return s[-amount]
def standardize_ssn(ssn):
    try:
        ssn = ssn.replace("-", "")
        ssn = "".join(ssn.split()) # Why not use ssn = ssn.replace("-", "") and that's it
        if (len(ssn) < 9 and ssn != "Missing"):
            ssn = "000000000" + ssn
            ssn = right(ssn, 9)
    except:
        pass
    return ssn
df.ssn = df.ssn.apply(standardize_ssn)
df
```

```python
# binning data
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# create the bin dividers
bins = [0, 60, 70, 80, 90, 100]
# create names for the four groups
group_names = ['F', 'D', 'C', 'B', 'A'] # Notice that there is one more bin value than there are group_names. This is because there needs to be a top and bottom limit for each bin.
df['lettergrade'] = pd.cut(df['grade'], bins, labels = group_names)
df
```

```python
# count number of observations
pd.value_counts(df['lettergrade'])
```
