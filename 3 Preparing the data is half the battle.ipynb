```python
# Cleaning Data

# Calculating and Removing Outliers
'''
    How to detect outliers:
        - Standard Deviations: If the data is normally distributed, then 95 percent of the data is within 1.96 standard deviations of the mean.
        - Interquartile Range (IQR): The IQR is the difference between the 25 percent quantile and the 75 percent quantile.
'''
```

```python
# Standard Deviations
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
meangrade = df['grade'].mean()
stdgrade = df['grade'].std()
toprange = meangrade + stdgrade * 1.96
botrange = meangrade - stdgrade * 1.96
copydf = df
copydf = copydf.drop(copydf[copydf['grade'] > toprange].index)
copydf = copydf.drop(copydf[copydf['grade'] < botrange].index)
copydf
```

```python
# Interquartile Range
import pandas as pd
Location = 'datasets/gradedata.csv'
df = pd.read_csv(Location)
q1 = df['grade'].quantile(.25)
q3 = df['grade'].quantile(.75)
iqr = q3 - q1
toprange = q3 + iqr * 1.5
botrange = q1 - iqr * 1.5
copydf = df
copydf = copydf.drop(copydf[copydf['grade'] > toprange].index)
copydf = copydf.drop(copydf[copydf['grade'] < botrange].index)
copydf
```

```python
# Missing Data in Pandas Dataframes
import pandas as pd
df = pd.read_csv("datasets/gradedatamissing.csv")
df.head()
```

```python
# drop rows with missing data

df_no_missing = df.dropna()
df_no_missing
```

```python
# add a column with empty values
import numpy as np
df['newcol'] = np.nan
df.head()
```

```python
# drop completely empty columns
df.dropna(axis=1, how='all')
```

```python
# replace empty cells with 0
df.fillna(0)
```

```python
# replace empty cells with average of column
df['grade'].fillna(df['grade'].mean(), inplace = True) # inplace = True means that the changes are saved to the dataframe right away
df
```

```python
# fill in missing grades with each gender's mean value of grade
df["grade"].fillna(df.groupby("gender")["grade"].transform("mean"), inplace = True)
```

```python
# selecting rows with no missing age or gender
df_select = df[df['age'].notnull() & df['gender'].notnull()]
df_select
```

```python
# Filtering inappropriate values
import pandas as pd
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel']
grades = [76, -2, 77, 78, 101]

GradeList = zip(names, grades)
df = pd.DataFrame(data = GradeList, columns = ['Names', 'Grades'])
df
```

```python
# filtering out impossible grades
'''
    pandas.DataFrame.loc
        - Access a group of rows and columns by labels or a boolean array
        - NOTE: using [[]] returns a DataFrame
'''
df.loc[df['Grades'] <= 100]
```

```python
# changing impossible grades
df.loc[(df['Grades'] >= 100, 'Grades')] = 100
df
```

```python
# Finding Duplicate Rows
import pandas as pd
names = ['Jan', 'John', 'Bob', 'Jan', 'Mary', 'Jon', 'Mel', 'Mel']
grades = [95, 78, 76, 95, 77, 78, 99, 100]
GradeList = zip(names, grades)
df = pd.DataFrame(data = GradeList, columns = ['Names', 'Grades'])
df
```

```python
# displaying only duplicates in the dataframe
df.duplicated()
```

```python
# displaying dataset without duplicates
df.drop_duplicates()
```

```python
# drop rows with duplicate names, keeping the last observation (this can happen if someone does your survey or retakes an exam again)
df.drop_duplicates(['Names'], keep = 'last')
```
