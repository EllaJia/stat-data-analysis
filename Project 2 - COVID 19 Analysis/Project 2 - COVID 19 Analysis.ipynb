```python
def assign_season(x):
    if x in (3, 4, 5):
        return 'a.Spring'
    elif x in (6, 7, 8, 9):
        return 'b.Summer'
    elif x in (10, 11):
        return 'c.Fall'
    else: 
        return 'd.Winter'
    
df['season']=df['month'].apply(assign_season)
```
