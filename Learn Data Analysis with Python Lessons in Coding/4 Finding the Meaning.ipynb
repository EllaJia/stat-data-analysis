```python
# Computing Aggregate Statistics on Matching Rows
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]
GradeList = zip(names,grades,bsdegrees,msdegrees,phddegrees)
columns=['Names','Grades','BS','MS','PhD']
df = pd.DataFrame(data = GradeList, columns=columns)
df
```

```python
# code for computing aggregate statistics
df.loc[df['PhD'] == 0].count()
```

```python
# compute aggregate statistics on a particular column
df.loc[df['PhD']==0]['Grades'].mean()
```

```python
# Regression
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
'''
    R-squared:
        The percentage of the variation in the data that can be accounted for by the regression.
    p-value:
        The probability that the independent variable has no effect on the dependent variable
'''

import statsmodels.formula.api as sm
result = sm.ols(formula = 'grade ~ age + exercise + hours',data = df).fit()
result.summary()
```

```python
'''
    Because the p-value of age is 0.595, let's rerun the regression leaving out the age
'''
import statsmodels.formula.api as sm
result = sm.ols(formula='grade ~ exercise + hours', data = df).fit()
result.summary()
```

```python
# Regression without Intercept
result = sm.ols(formula = 'grade ~ age + exercise + hours - 1', data = df).fit()
result.summary()
```

```python
# Basic Pivot Table
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# get averages of all numeric columns categorized by gender
'''
    pivot_table
        it's smart enough to assume that we want the averages of all the numeric columns
'''
pd.pivot_table(df, index=['gender'])
```

```python
# average grade by gender
pd.pivot_table(df, values=['grade'], index = ['gender'])
```

```python
# minimum grade by gender
pd.pivot_table(df, values = ['grade'], index = ['gender'], aggfunc = 'min')
```

```python
# max grade by gender and age
pd.pivot_table(df, index=['gender', 'age'], aggfunc = 'max', values = ['hours'])
```

```python
# average grade and hours by gender
pd.pivot_table(df, index = ['gender'], aggfunc = 'mean', values = ['grade', 'hours'])
```

```python
# average grade and hours by gender
df2 = df.loc[df['age'] == 17]
pd.pivot_table(df2, index = ['gender'], aggfunc = 'mean', values = ['grade', 'hours'])
```

```python
# average grade and hours by gender, include totals on pivot tables
df2 = df.loc[df['age'] == 17]
pd.pivot_table(df2, index = ['gender'], aggfunc = 'mean', values = ['grade', 'hours'], margins = 'True')
```
