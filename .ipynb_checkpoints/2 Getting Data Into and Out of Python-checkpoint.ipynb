```python
'''
Additional materials for zip function:
    zip() returns an iterator of tuples.
    
    Properties:
        zip() is lazy. The elements won't be processed until the iterable is iterated on, e.g. by a for loop or by wrapping in a list.
        Option:
            strict=True: check the lengths of iterables are identical. Without it, any bug that results in iterables of different lengths will be silenced.
'''
for item in zip([1,2,3], ['sugar', 'spice', 'everything nice']):
    print(item)
```

```python
#  zip list
list(zip(('a', 'b', 'c'), ))
```

```python
# zip enumerate
names = ['Mukesh', 'Roni', 'Chari']
ages = [24, 50, 18]

for i, (name, age) in enumerate(zip(names, ages)):
    print(i, name, age)
```

```python
# zip dictionary
stocks = ['reliance', 'infosys', 'tcs']
prices = [2175, 1127, 2750]

new_dict = dict(zip(stocks, prices))

'''
Another syntax for dictionary comprehension is:
    dictionary = {key: value for vars in iterable}
    os the codes above can also be written as:
        new_dict = {stocks: prices for stocks, prices in zip(stocks, prices)}
'''
print(new_dict)
```

```python
# unzipping values

# initializing lists
name = ["Manjeet", "Nikhil", "Shambhavi", "Astha"]
roll_no = [4, 1, 3, 2]
marks = [40, 50, 60, 70]

mapped = zip(name, roll_no, marks)
mapped = list(mapped)

print("The zipped result is : ", end="")
print(mapped)
print("\n")
 
# unzipping values (the result would be tuples)
namz, roll_noz, marksz = zip(*mapped)
 
print("The unzipped result: \n", end="")
print("The name list is : ", end="")
print(namz)
print("The roll_no list is : ", end="")
print(roll_noz)
print("The marks list is : ", end="")
print(marksz)
```

```python
import pandas as pd
names = ['Bob', 'Jessica', 'Mary',' John', 'Mel']
grades = [76, 95, 77, 78, 99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]
Degrees = zip(names, grades, bsdegrees, msdegrees, phddegrees)
column = ['Names', 'Grades', 'BS', 'MS', 'PhD']
df = pd.DataFrame(data = Degrees, columns = column)
df.to_csv('practice_to_csv.csv', index = False)
```

```python
# Saving Data to Excel Files

import pandas as pd
names = ['Bob', 'Jessica', 'Mary', 'John',' Mel']
grades = [76, 95, 77, 78, 99]
GradeList = zip(names, grades)
df = pd.DataFrame(data = GradeList, columns = ['Names', 'Grades'])

'''
    pd.ExcelWriter():
        class for writing DataFrame objects into excel sheets
    DataFrame.to_excel():
        a function from pandas
'''
writer = pd.ExcelWriter('dataframe.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.save()
```

```python
# Exporting Multiple Dataframes to different sheets of Excel
names = ['Bob', 'Jessica', 'Mary', 'John',' Mel']
id = [1,2,3,4,5]
IdList = zip(names, id)
df2 = pd.DataFrame(data = IdList, columns = ['Names', 'ID'])

writer = pd.ExcelWriter('dataframe2.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1')
df2.to_excel(writer, sheet_name='Sheet2', index=False)
writer.save()
```

```python
# Combining Data from Multiple Excel Files
import pandas as pd

'''
    DataFrame.append(other, ignore_index=False, ...):
        a function from pandas. Append rows of other to the end of calller, returning a new DataFrame. Columns in other that are not in the caller are added as new columns.
    DataFrame.describe():
        a function from pandas. It's used to generate descriptive statistics.
'''
# each of the original data1/2/3.xlsx has 100 rows
all_data = pd.DataFrame()
df = pd.read_excel("datasets/data1.xlsx")
all_data = all_data.append(df, ignore_index=True) # length 100

df = pd.read_excel("datasets/data2.xlsx")
all_data = all_data.append(df, ignore_index=True) # length 200

df = pd.read_excel("datasets/data3.xlsx")
all_data = all_data.append(df, ignore_index=True) # length 300

all_data.describe()
```

```python
# Loop through all the files in a directory
import pandas as pd
import glob

'''
    glob module:
        It finds all the pathnames matching a specified pattern according to the rules used by the Unix shell.
'''
# Way 1: glob module
all_data = pd.DataFrame()
for f in glob.glob("datasets/data*.xlsx"):
    df = pd.read_excel(f)
    all_data = all_data.append(df, ignore_index=True)
all_data.describe()
```

```python
# Way 2: os module
import os

all_data = pd.DataFrame()
for f in os.listdir("datasets"):
    if f.startswith("data") and f.endswith(".xlsx"):
        filename = os.path.join('datasets', f) # Notice: Our path is still outside of ".../datasets"
        ""
        df = pd.read_excel(filename)
        all_data = all_data.append(df, ignore_index=True)
all_data.describe()
# print(all_data.shape)
```
